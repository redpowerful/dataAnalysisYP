Средние пользовательские рейтинги жанров Action (англ. «действие») и Sports (англ. «виды спорта») разные.  
Задайте самостоятельно пороговое значение alpha.  
Поясните:  
1) Как вы сформулировали нулевую и альтернативную гипотезы;  
2) Какой критерий применили для проверки гипотез и почему.

Н0 (нулевая гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.  
Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.  

<div class="alert alert-success">
<font size="5"><b>Комментарий ревьюера</b></font>

Импорт необходимых библиотек следует производить в первой ячейке работы, как то рекомендует стандарт.

```python

xone_ratings = games[(games['platform'] == 'XOne') & (games['user_score'] > 0)]['user_score']
pc_ratings = games[(games['platform'] == 'PC') & (games['user_score'] > 0)]['user_score']
stat, p = ttest_ind(xone_ratings, pc_ratings)

print(f't = {stat} \np-value = {p}')
print('Пороговое значение alpha = 0.05')
```

Нулевая гипотеза отклоняется, следовательно: средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой - верное утверждение

```python
print('Cредние пользовательские рейтинги PC - {:.2f}'.format(pc_ratings.mean()))
print('Cредние пользовательские рейтинги Xbox One  - {:.2f}'.format(xone_ratings.mean()))
```

Рейтинги на PC выше чем на Xbox
